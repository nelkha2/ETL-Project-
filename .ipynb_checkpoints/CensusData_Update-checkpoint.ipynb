```python
# Dependencies
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
import requests
from census import Census
from sqlalchemy import create_engine
import seaborn as sns

# Census & gmaps API Keys
api_key = "f342c203f985f826ddbfc242fce4a634199dbcd5"
c = Census(api_key, year=2017)
```

```python
#Store CSV into Dataframe
csv_file = "stateexportytd.csv"
state_export = pd.read_csv(csv_file)
state_export_df = state_export.rename(columns={"State":"state",
                                               "Man_Total(m)": "man_total_million",
                                               "Man_%_of_Total":"man_percentage_of_total",
                                               "NonMan_Total(m)":"non_man_total_million",
                                               "NonMan_%_of_Total":"non_man_percentage_of_total",
                                               "Reexport Total(m)":"reexport_total_millin",
                                               "Total_Export_YTD(m)":"total_expert_ytd_million"})
state_export_df.head()
```

```python
# Run Census Search to retrieve data on all states
# Note the addition of "B23025_005E" for unemployment count
census_data = c.acs5.get(("NAME", "B19013_001E", "B01003_001E", "B01002_001E",
                          "B19301_001E",
                          "B17001_002E",
                          "B23025_005E"), {'for': 'state:*'})
# Convert to DataFrame
census_pd = pd.DataFrame(census_data)
# Column Reordering
census_pd = census_pd.rename(columns={"B01003_001E": "Population",
                                      "B01002_001E": "Median Age",
                                      "B19013_001E": "Household Income",
                                      "B19301_001E": "Per Capita Income",
                                      "B17001_002E": "Poverty Count",
                                      "B23025_005E": "Unemployment Count",
                                      "NAME": "Name", "state": "State"})
census_pd.head()
```

```python
# Male population data for 15-44yr olds
male_age_data = c.acs5.get(("NAME", "B01001_006E","B01001_007E","B01001_008E","B01001_009E","B01001_010E","B01001_011E",
                     "B01001_012E","B01001_013E","B01001_014E","B01001_015E","B01001_016E","B01001_017E","B01001_018E","B01001_019E","B01001_020E"), {'for': 'state:*'})
# Female population data for 15-44yr olds
female_age_data = c.acs5.get(("NAME", "B01001_030E","B01001_031E","B01001_032E","B01001_033E","B01001_034E","B01001_035E",
                     "B01001_036E","B01001_037E","B01001_038E","B01001_039E", "B01001_040E", "B01001_041E", "B01001_042E", "B01001_043E", "B01001_044E" ), {'for': 'state:*'})

# Convert to DataFrame
male_pd = pd.DataFrame(male_age_data)
male_pd= male_pd.rename(columns={"state": "State"})
female_pd= pd.DataFrame(female_age_data)
female_pd= female_pd.rename(columns={"state": "State"})

male_pd.head()
female_pd.head()
```

```python
# Total population by gender
male_pd["Total_Male"]= male_pd.B01001_006E + male_pd.B01001_007E + male_pd.B01001_008E + male_pd.B01001_009E \
   + male_pd.B01001_010E + male_pd.B01001_011E + male_pd.B01001_012E + male_pd.B01001_013E + male_pd.B01001_014E + male_pd.B01001_015E\
   + male_pd.B01001_016E + male_pd.B01001_017E + male_pd.B01001_018E + male_pd.B01001_019E + male_pd.B01001_020E
female_pd["Total_Female"]= female_pd.B01001_030E + female_pd.B01001_031E + female_pd.B01001_032E + female_pd.B01001_033E \
   + female_pd.B01001_034E + female_pd.B01001_035E + female_pd.B01001_036E + female_pd.B01001_037E + female_pd.B01001_038E\
   + female_pd.B01001_039E + female_pd.B01001_040E + female_pd.B01001_041E + female_pd.B01001_042E + female_pd.B01001_043E\
   + female_pd.B01001_044E
```

```python
male_pd.head()
```

```python
gender_merge= pd.merge(male_pd,female_pd, on="State")
gender_merge.head()
```

```python
census_df = gender_merge[["NAME_y", "Total_Female", "Total_Male"]]

census_df["Working_Pop"] = census_df["Total_Female"] + census_df["Total_Male"]
census_df.head()
```

```python
census_df1 = census_df.rename(columns={"NAME_y": "Name"})
census_df1.head()
```

```python
census_merge= pd.merge(census_df1,census_pd, on="Name")
census_merge = census_merge.rename(columns={"Name":"name",
                                            "Total_Female":"total_female",
                                            "Total_Male":"total_male",
                                            "Working_Pop":"working_pop",
                                            "Median Age":"median_age",
                                            "Population":"population",
                                            "Poverty Count":"poverty_count",
                                            "Household Income":"household_income",
                                            "Per Capita Income":"per_capita_income",
                                            "Unemployment Count":"unemployment_count",
                                            "State":"state_id"
                                           })
census_merge.head()
```

```python
#Connect to local database
rds_connection_string = "postgres:postgres@localhost:5432/etl_project"
engine = create_engine(f'postgresql://{rds_connection_string}')
```

```python
#Check Table Names
engine.table_names()
```

```python
#load CSV Data into PostgresSQL Local DB
state_export_df.to_sql(name='state_export', con=engine, if_exists='append', index=False)
```

```python
#Load API DATAFRAME into PostgresSQL Local DB
census_merge.to_sql(name='census_data', con=engine, if_exists='append', index=False)
```

```python
plotdf= pd.read_sql_query('select distinct census_data.name, census_data.working_pop, state_export.total_expert_ytd_million from census_data join state_export on census_data.name=state_export.state order by census_data.working_pop desc ',con=engine)
plotdf.head()
```

```python
state_name = plotdf["name"]
working_pop = plotdf["working_pop"]
sales_export= plotdf["total_expert_ytd_million"]
plt.figure(figsize=(20,10))
sns.set(style="whitegrid")
tips = sns.load_dataset("tips")
sns.barplot(x= state_name , y= working_pop, data=tips)
sns.lineplot(x= state_name, y= sales_export, data=tips)
plt.xticks(rotation=90)
plt.title("Working Population by State")
plt.xlabel("State")
plt.ylabel("Working Population (millions)")
plt.show()
```

```python
#State vs Export Revenue
plotdf2= pd.read_sql_query('select distinct census_data.name, census_data.working_pop, state_export.total_expert_ytd_million from census_data join state_export on census_data.name=state_export.state order by state_export.total_expert_ytd_million desc ',con=engine)
x_axis = plotdf2['name']
y_axis = plotdf2['total_expert_ytd_million']
plt.figure(figsize=(20,10))
sns.set(style="whitegrid")
tips = sns.load_dataset("tips")
sns.lineplot(x= x_axis , y= y_axis, data=tips, sort= False)
#graph_state_export, = plt.plot(x_axis, y_axis, color="blue", label="State Export Revenue" )
plt.title('Export Revenue by State')
plt.xlabel('State')
plt.ylabel('Export Revenue')
plt.xticks(rotation=90)
plt.show()
```
